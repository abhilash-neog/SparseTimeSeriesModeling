```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
etth1_path="/raid/abhilash/forecasting_datasets/ETT/ETTh1.csv"
etth2_path="/raid/abhilash/forecasting_datasets/ETT/ETTh2.csv"
etth1=pd.read_csv(etth1_path)
etth2=pd.read_csv(etth2_path)
```

```python
for feat in etth1.columns[1:]:
    avg_h1=etth1[feat].mean()
    min_h1=etth1[feat].min()
    max_h1=etth1[feat].max()
    
    avg_h2=etth2[feat].mean()
    min_h2=etth2[feat].min()
    max_h2=etth2[feat].max()
    
    print(f"\n\nFeature = {feat} :: \nETTh1 avg={avg_h1}\nETTh2 avg={avg_h2}\n\nETTh1 min={min_h1}\nETTh2 min={min_h2}\n\nETTh1 max={max_h1}\nETTh2 max={max_h2}")
```

## ZOOMED IN

```python
for feat in etth1.columns[1:]:
    fig, axes = plt.subplots(1,2, figsize=(15, 5))
    axes[0].plot(etth1[feat].values[:200], label='etth1_'+str(feat))
    axes[0].legend()
    axes[0].grid()
    axes[1].plot(etth2[feat].values[:200], label='etth2_'+str(feat))
    axes[1].legend()
    axes[1].grid()
```
