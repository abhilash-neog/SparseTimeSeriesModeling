```python
import numpy as np


# Example predictions and true values
pred = np.array([1, 2, 3])
true = np.array([np.nan, 1, np.nan])

mask = ~np.isnan(true)  # Create a mask to ignore NaN values in true
print(mask)
pred_filtered = pred[mask]  # Filter the predictions using the mask
true_filtered = true[mask]  # Filter the true values using the mask

mse_value=  np.mean((pred_filtered - true_filtered) ** 2)  # Calculate MSE
# Output the result
print("MSE:", mse_value)
```
