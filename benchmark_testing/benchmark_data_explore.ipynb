```python
def generate_splits(df, config_name):
    start_date = df.index.min()
    
    if config_name in ["m1", "m2"]:
        factor = 4  # 15-min frequency
    else:
        factor = 1  # hourly frequency
        
    train_end_date_index = 12 * 30 * 24 * factor  # 1 year

    dev_end_date_index = train_end_date_index + 4 * 30 * 24 * factor  # 1 year + 4 months
    test_end_date_index = train_end_date_index + 8 * 30 * 24 * factor  # 1 year + 8 months

    return train_end_date_index, dev_end_date_index, test_end_date_index
```

```python
etth1_train_ind, etth1_dev_ind, etth1_test_ind = generate_splits(etth1, "h1")
```

```python
train = etth1.iloc[:etth1_train_ind,:]
dev = etth1.iloc[etth1_train_ind:etth1_dev_ind, :]
test = etth1.iloc[etth1_dev_ind:etth1_test_ind, :]
```
